```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/exchange_rate.csv')

# Convert 'date' column to datetime objects and set it as index
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y %H:%M')
df.set_index('date', inplace=True)

# Check the first few rows of the data and available columns
print(df.head())
print(df.columns)
```

```python
# Plot the time series for USD to AUD exchange rate
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Ex_rate'], label='USD to AUD Exchange Rate')
plt.title('USD to Australian Dollar Exchange Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Fill missing values using linear interpolation
df['Ex_rate'] = df['Ex_rate'].interpolate(method='linear')

# Check if there are any missing values after interpolation
print(df.isnull().sum())
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot ACF and PACF
plt.figure(figsize=(12, 6))
plt.subplot(121)
plot_acf(df['Ex_rate'], lags=40, ax=plt.gca())
plt.subplot(122)
plot_pacf(df['Ex_rate'], lags=40, ax=plt.gca())
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit the ARIMA model (p, d, q) = (1, 1, 1) based on ACF/PACF insights
arima_model = ARIMA(df['Ex_rate'], order=(1, 1, 1))
arima_fitted = arima_model.fit()

# Display the ARIMA model summary
print(arima_fitted.summary())
```

```python
# Plot residuals of the ARIMA model
residuals = arima_fitted.resid

# Plot residuals
plt.figure(figsize=(10, 6))
plt.plot(residuals)
plt.title('ARIMA Model Residuals')
plt.xlabel('Date')
plt.ylabel('Residuals')
plt.grid(True)
plt.show()

# Check the ACF of residuals
plot_acf(residuals, lags=40)
plt.show()
```

```python
# Forecast the next 30 days
forecast_steps = 30
arima_forecast = arima_fitted.forecast(steps=forecast_steps)

# Plot the forecasted values against the actual values
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Ex_rate'], label='Actual Data')
plt.plot(pd.date_range(df.index[-1], periods=forecast_steps + 1, freq='D')[1:], arima_forecast, color='red', label='ARIMA Forecast')
plt.title('ARIMA Model Forecast')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Fit Holt’s Linear Trend model
holt_model = ExponentialSmoothing(df['Ex_rate'], trend='add', seasonal=None)
holt_fitted = holt_model.fit()

# Summary of the fitted model
print(holt_fitted.summary())
```

```python
# choosing the one with the lowest AIC
holt_model_optimized = ExponentialSmoothing(df['Ex_rate'], trend='add', seasonal=None).fit()
print('AIC for Holt’s model:', holt_model_optimized.aic)
```

```python
# Forecasting with Holt's Linear Trend model
holt_forecast = holt_fitted.forecast(steps=30)

# Plot the forecasted values against the actual values
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Ex_rate'], label='Actual Data')
plt.plot(pd.date_range(df.index[-1], periods=forecast_steps + 1, freq='D')[1:], holt_forecast, color='green', label='Holt’s Linear Forecast')
plt.title('Holt’s Linear Trend Model Forecast')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Computing error metrics for ARIMA model
mae_arima = mean_absolute_error(df['Ex_rate'][-forecast_steps:], arima_forecast)
rmse_arima = np.sqrt(mean_squared_error(df['Ex_rate'][-forecast_steps:], arima_forecast))
mape_arima = np.mean(np.abs((df['Ex_rate'][-forecast_steps:] - arima_forecast) / df['Ex_rate'][-forecast_steps:])) * 100

# Computing error metrics for Holt’s model
mae_holt = mean_absolute_error(df['Ex_rate'][-forecast_steps:], holt_forecast)
rmse_holt = np.sqrt(mean_squared_error(df['Ex_rate'][-forecast_steps:], holt_forecast))
mape_holt = np.mean(np.abs((df['Ex_rate'][-forecast_steps:] - holt_forecast) / df['Ex_rate'][-forecast_steps:])) * 100

print(f"ARIMA Model - MAE: {mae_arima}, RMSE: {rmse_arima}, MAPE: {mape_arima}")
print(f"Holt’s Model - MAE: {mae_holt}, RMSE: {rmse_holt}, MAPE: {mape_holt}")
```
