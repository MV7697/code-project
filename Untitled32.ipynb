```python
import pandas as pd

# Load the dataset (replace 'anime.csv' with your dataset file path)
df = pd.read_csv('/content/anime.csv')

# Show the first few rows of the dataset
df.head()
```

```python
# Check for missing values
df.isnull().sum()

# Check column names and basic info
df.info()

# Describe the dataset
df.describe()
```

```python
# Example: Fill missing values with the median or mean
df['rating'].fillna(df['rating'].median(), inplace=True)
```

```python
# Example: One-hot encoding for genres
genres_encoded = df['genre'].str.get_dummies(sep=', ')
df = pd.concat([df, genres_encoded], axis=1)

# Drop original genre column (optional)
df.drop('genre', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['rating', 'episodes']] = scaler.fit_transform(df[['rating', 'episodes']])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Assume 'df' has numerical values for each anime, including one-hot encoded genres, ratings, and episodes
# Get a list of existing columns
existing_columns = df.columns

# Check if 'anime_title', 'name', and 'type' are in the columns (removed 'unique_id')
columns_to_drop = ['anime_title', 'name', 'type']
columns_to_drop = [col for col in columns_to_drop if col in existing_columns]

# Drop only the existing columns
similarity_matrix = cosine_similarity(df.drop(columns=columns_to_drop))
```

```python
from sklearn.model_selection import train_test_split

# Assuming we have ratings data
X_train, X_test = train_test_split(df, test_size=0.2, random_state=42)
```
