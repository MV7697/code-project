```python
import numpy as np
from scipy.stats import chi2_contingency

# Function to compute Cramr's V
def cramers_v(x, y):
    contingency = pd.crosstab(x, y)
    chi2, p, dof, ex = chi2_contingency(contingency)
    n = contingency.sum().sum()
    return np.sqrt(chi2 / (n * (min(contingency.shape) - 1)))

# Calculate Cramr's V for all pairs of features
cramers_matrix = np.zeros((df.shape[1], df.shape[1]))
for i in range(df.shape[1]):
    for j in range(df.shape[1]):
        cramers_matrix[i, j] = cramers_v(df.iloc[:, i], df.iloc[:, j])

# Visualizing the Cramr's V matrix
plt.figure(figsize=(10, 8))
sns.heatmap(cramers_matrix, annot=True, cmap='coolwarm', xticklabels=df.columns, yticklabels=df.columns)
plt.title("Cramr's V Correlation Matrix")
plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# Define different kernel types
kernels = ['linear', 'poly', 'rbf']
results = {}

# Train and evaluate the SVM with different kernels
for kernel in kernels:
    # Initialize the SVM model with the specific kernel
    svm_model = SVC(kernel=kernel)

    # Train the model on the training data
    svm_model.fit(X_train, y_train)

    # Make predictions on the test data
    y_pred = svm_model.predict(X_test)

    # Evaluate the model using classification report
    results[kernel] = classification_report(y_test, y_pred, output_dict=True)

# Display the classification reports for each kernel
for kernel in kernels:
    print(f"Classification Report for Kernel: {kernel}")
    print(results[kernel]['accuracy'])  # Print accuracy for comparison
    print(classification_report(y_test, svm_model.predict(X_test)))
```

```python
accuracies = [results[kernel]['accuracy'] for kernel in kernels]
precisions = [results[kernel]['weighted avg']['precision'] for kernel in kernels]
recalls = [results[kernel]['weighted avg']['recall'] for kernel in kernels]
f1_scores = [results[kernel]['weighted avg']['f1-score'] for kernel in kernels]

# Create a dataframe for easy plotting
import pandas as pd

df_results = pd.DataFrame({
    'Kernel': kernels,
    'Accuracy': accuracies,
    'Precision': precisions,
    'Recall': recalls,
    'F1-score': f1_scores
})

# Plot the results
df_results.set_index('Kernel').plot(kind='bar', figsize=(10, 6))
plt.title('SVM Kernel Performance Comparison')
plt.xlabel('Kernel Type')
plt.ylabel('Score')
plt.xticks(rotation=0)
plt.show()
```
