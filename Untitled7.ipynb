```python
from sklearn.preprocessing import PolynomialFeatures

# Transform features for polynomial regression (degree=2)
poly = PolynomialFeatures(degree=2)
X_poly_train = poly.fit_transform(X_train)
X_poly_test = poly.transform(X_test)

# Initialize and train polynomial regression model
poly_model = LinearRegression()
poly_model.fit(X_poly_train, y_train)

# Predict and evaluate
y_poly_pred = poly_model.predict(X_poly_test)
```

```python
# Initialize Ridge Regression model with alpha (regularization strength)
ridge_model = Ridge(alpha=1.0)

# Train the model
ridge_model.fit(X_train, y_train)

# Predict and evaluate
y_ridge_pred = ridge_model.predict(X_test)
```

```python
def evaluate_model(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    print(f"MAE: {mae}")
    print(f"MSE: {mse}")
    print(f"R: {r2}")

# Evaluate Linear Regression Model
print("Linear Regression Model Evaluation:")
evaluate_model(y_test, y_pred)

# Evaluate Polynomial Regression Model
print("\nPolynomial Regression Model Evaluation:")
evaluate_model(y_test, y_poly_pred)

# Evaluate Ridge Regression Model
print("\nRidge Regression Model Evaluation:")
evaluate_model(y_test, y_ridge_pred)
```

```python
# Initialize Lasso Regression model
lasso_model = Lasso(alpha=1.0)

# Train the model
lasso_model.fit(X_train, y_train)

# Predict and evaluate
y_lasso_pred = lasso_model.predict(X_test)

# Evaluate the Lasso model
print("\nLasso Regression Model Evaluation:")
evaluate_model(y_test, y_lasso_pred)
```
