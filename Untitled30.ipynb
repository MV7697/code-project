```python
import pandas as pd

# Load dataset
df = pd.read_csv("/content/EastWestAirlines.csv")

# Check for missing values
print(df.isnull().sum())

# Impute missing values (for numerical columns)
df.fillna(df.mean(), inplace=True)

# Or drop missing values
df.dropna(inplace=True)
```

```python
import pandas as pd
from scipy import stats
import numpy as np # Import the numpy library

# Z-score method
df_no_outliers = df[(np.abs(stats.zscore(df.select_dtypes(include=['float64', 'int64']))) < 3).all(axis=1)]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_df = scaler.fit_transform(df.select_dtypes(include=['float64', 'int64']))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize distribution of data
sns.pairplot(df)

# Visualize box plots for detecting outliers
sns.boxplot(data=df)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Apply the Elbow Method to find the optimal K
distortions = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_df)
    distortions.append(kmeans.inertia_)

# Plot the Elbow curve
plt.plot(range(1, 11), distortions)
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()

# Apply K-Means with the optimal k
optimal_k = 3  # Change according to the Elbow plot
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(scaled_df)

# Add cluster labels to the dataframe
df['KMeans_Cluster'] = kmeans.labels_
```

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch

# Dendrogram to choose the number of clusters
plt.figure(figsize=(10, 7))
sch.dendrogram(sch.linkage(scaled_df, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Data points')
plt.ylabel('Euclidean distances')
plt.show()

# Apply Agglomerative Clustering
agglomerative = AgglomerativeClustering(n_clusters=3, linkage='ward')
df['Hierarchical_Cluster'] = agglomerative.fit_predict(scaled_df)
```

```python
from sklearn.cluster import DBSCAN

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
df['DBSCAN_Cluster'] = dbscan.fit_predict(scaled_df)
```

```python
# Cluster summary for K-Means
print(df.groupby('KMeans_Cluster').mean())

# Cluster summary for Hierarchical
print(df.groupby('Hierarchical_Cluster').mean())

# Cluster summary for DBSCAN
print(df.groupby('DBSCAN_Cluster').mean())
```

```python
# Plot K-Means clusters
plt.scatter(df['Balance'], df['Bonus_miles'], c=df['KMeans_Cluster'], cmap='viridis')
plt.title('K-Means Clusters')
plt.xlabel('Balance')  # Update x-axis label
plt.ylabel('Bonus_miles')  # Update y-axis label
plt.show()

# Plot Hierarchical clusters
# Replace 'feature1' and 'feature2' with the actual column names from your DataFrame
plt.scatter(df['Balance'], df['Bonus_miles'], c=df['Hierarchical_Cluster'], cmap='viridis')
plt.title('Hierarchical Clusters')
plt.xlabel('Balance')  # Update x-axis label
plt.ylabel('Bonus_miles')  # Update y-axis label
plt.show()

# Plot DBSCAN clusters
# Replace 'feature1' and 'feature2' with the actual column names from your DataFrame
plt.scatter(df['Balance'], df['Bonus_miles'], c=df['DBSCAN_Cluster'], cmap='viridis')
plt.title('DBSCAN Clusters')
plt.xlabel('Balance')  # Update x-axis label
plt.ylabel('Bonus_miles')  # Update y-axis label
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Silhouette score for K-Means
kmeans_silhouette = silhouette_score(scaled_df, df['KMeans_Cluster'])
print(f'K-Means Silhouette Score: {kmeans_silhouette}')

# Silhouette score for DBSCAN
dbscan_silhouette = silhouette_score(scaled_df, df['DBSCAN_Cluster'])
print(f'DBSCAN Silhouette Score: {dbscan_silhouette}')
```
